```python
#W8A3:
try:
    a = int(input("Nhập số nguyên dương a: "))
    if a <= 0:
        raise ValueError

    seen = set()

    while a != 1:
        if a in seen:
            print("Không phải số hạnh phúc")
            break

        seen.add(a)

        s = 0
        while a > 0:
            digit = a % 10
            s += digit * digit
            a //= 10

        a = s

    if a == 1:
        print("Là số hạnh phúc")

except ValueError:
    print("Lỗi nhập liệu: vui lòng nhập số nguyên dương")

```

```python
#W8A5:
lst = input("Nhập các phần tử, cách nhau bằng dấu cách: ").split()
new_list = []

for x in lst:
    try:
        new_list.append(int(x))
    except ValueError:
        print(f"Cảnh báo: '{x}' không thể chuyển sang int, bỏ qua.")

print("List sau khi chuyển:", new_list)
```

```python
#W10A10:
try:
    N = int(input())
    if N <= 0:
        raise ValueError

    arr = list(map(int, input().split()))
    if len(arr) != N:
        raise ValueError

    if len(set(arr)) != N:
        print("Mang khong hop le")
    else:
        print(sum(arr))

except:
    print("Mang khong hop le")
```
